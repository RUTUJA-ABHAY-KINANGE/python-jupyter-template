### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Data for each plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Line chart
axs[0, 0].plot(x, y, marker='o', color='b')
axs[0, 0].set_title('Line Chart')

# Second subplot: Bar chart
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Third subplot: Scatter plot
axs[1, 0].scatter(x, y, color='orange')
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```
