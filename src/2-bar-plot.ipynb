```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Bar plot with different colors
colors = ['red', 'green', 'blue', 'purple']
plt.bar(categories, values, color=colors)

# Add title
plt.title('Bar Plot with Colors')

# Show plot
plt.show()
```
